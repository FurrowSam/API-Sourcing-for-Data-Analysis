```python
from bs4 import BeautifulSoup
import requests

# URL of the webpage to scrape
url = "https://www.pendulummag.com/health-wellness/polish-whiteborscht-chefmichaelangeloni"

# Send an HTTP request
response = requests.get(url)

if response.status_code == 200:
    # Parse the page with BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")

    # Extract the title
    title = soup.find("h1").text.strip() if soup.find("h1") else "No title found"

    # Initialize flags
    ingredients_start = False
    instructions_start = False

    # Storage for ingredients and instructions
    ingredients = []
    instructions = []

    # Find all relevant <p> and <li> tags
    content = soup.find_all(["p", "li"])

    for element in content:
        text = element.text.strip()

        # Start and stop conditions for ingredients
        if "ingredients" in text.lower():
            ingredients_start = True
            instructions_start = False
            continue
        elif "instructions" in text.lower():
            ingredients_start = False
            instructions_start = True
            continue

        # Append text to ingredients or instructions based on flags
        if ingredients_start and text:
            ingredients.append(text)
        elif instructions_start and text:
            instructions.append(text)

    # Filter out unrelated content
    ingredients = [
        item for item in ingredients
        if any(char.isdigit() or char.isalpha() for char in item)  # Ensure it's not an empty or unrelated line
    ]
    instructions = [
        step for step in instructions
        if len(step) > 10  # Exclude unrelated or too-short text
    ]

    # Print the recipe details
    print(f"Title: {title}\n")

    print("Ingredients:\n")
    for ingredient in ingredients:
        print(f"- {ingredient}")

    print("\nInstructions:\n")
    for step, instruction in enumerate(instructions, 1):
        print(f"Step {step}: {instruction}")

else:
    print(f"Failed to fetch the page. Status code: {response.status_code}")
```
