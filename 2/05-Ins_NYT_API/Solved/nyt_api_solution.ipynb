```python
# Import dependencies
import requests
import json
from dotenv import load_dotenv
import os
```

```python
# Load environment variables and New York Times API key
load_dotenv("/Users/18176/mykeys.env")
api_key = os.getenv("NYT_API_KEY")
type(api_key)
```

```python
# New York Times Article API URL
url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"

# Search for articles that mention granola
query = "borscht"
```

```python
# Build query URL
query_url = url + "api-key=" + api_key + "&q=" + query
```

```python
# Request articles
articles = requests.get(query_url).json()

# The "response" property in articles contains the actual articles
# list comprehension.
articles_list = articles["response"]["docs"]
print(json.dumps(articles_list, indent=2))
```

```python
# Print the web_url of each stored article
print("Your Reading List")
for article in articles_list:
    print(article["web_url"])
```

```python
# Send a request to the New York Times Article Search API
response = requests.get(query_url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()
    
    # Extract articles
    articles = data.get("response", {}).get("docs", [])
    
    # Loop through articles and print their headlines and URLs
    for article in articles:
        headline = article.get("headline", {}).get("main", "No headline")
        web_url = article.get("web_url", "No URL")
        print(f"Headline: {headline}")
        print(f"Read here: {web_url}\n")
else:
    print(f"Failed to fetch articles. Status code: {response.status_code}")
```

```python
from bs4 import BeautifulSoup
import requests

# URL of the webpage to scrape
url = "https://www.pendulummag.com/health-wellness/polish-whiteborscht-chefmichaelangeloni"

# Send an HTTP request
response = requests.get(url)

if response.status_code == 200:
    # Parse the page with BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")

    # Debug: Print the full HTML content (optional, for inspection purposes)
    # print(soup.prettify())

    # Extract the title
    title = soup.find("h1").text.strip() if soup.find("h1") else "No title found"

    # Locate the ingredients section (Debug step)
    ingredients_section = soup.find_all("li")  # Update this based on inspection
    print("Ingredients Section Raw HTML:", ingredients_section)  # Debug

    # Process the ingredients
    ingredients = [item.text.strip() for item in ingredients_section]
    print("Extracted Ingredients:", ingredients)  # Debug

    # Locate the instructions section (Debug step)
    instructions_section = soup.find_all("p")  # Update this based on inspection
    print("Instructions Section Raw HTML:", instructions_section)  # Debug

    # Process the instructions
    instructions = [step.text.strip() for step in instructions_section]
    print("Extracted Instructions:", instructions)  # Debug

    # Print the recipe details
    print(f"Title: {title}\n")

    print("Ingredients:\n")
    for ingredient in ingredients:
        print(f"- {ingredient}")

    print("\nInstructions:\n")
    for step, instruction in enumerate(instructions, 1):
        print(f"Step {step}: {instruction}")

else:
    print(f"Failed to fetch the page. Status code: {response.status_code}")
```
